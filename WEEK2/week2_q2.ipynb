- **Omega Policy**: Push left if the pole velocity ω (state[3]) is negative, otherwise push right.

```python
# Define the function to run the environment for a fixed number of episodes with a given policy
def run_for_episodes_with_policy(env, num_episodes=100, policy='theta'):
    """
    Run the CartPole environment for a fixed number of episodes with a given policy.
    - The 'policy' can be either 'theta' or 'omega'.
    - The function returns the rewards per episode and the cumulative rewards.
    """
    rewards_per_episode = []  # List to store the total reward for each episode
    
    for episode in range(num_episodes):
        state = env.reset()  # Reset the environment at the start of each episode
        total_reward = 0  # Initialize the total reward for the current episode
        done = False  # Boolean flag to check if the episode is finished
        
        # Loop through each step within the episode
        while not done:
            if policy == 'theta':
                # Theta Policy: Push the cart left if θ < 0, otherwise push right
                action = 0 if state[2] < 0 else 1  # θ = state[2], action 0 = left, action 1 = right
            elif policy == 'omega':
                # Omega Policy: Push the cart left if ω < 0, otherwise push right
                action = 0 if state[3] < 0 else 1  # ω = state[3], action 0 = left, action 1 = right
            
            # Take the action and observe the new state and reward
            next_state, reward, done, info = env.step(action)  # Perform the action in the environment
            total_reward += reward  # Accumulate the reward for the episode
            state = next_state  # Update the state for the next step

        rewards_per_episode.append(total_reward)  # Store the total reward for the current episode
    
    # Calculate cumulative rewards (i.e., running total of rewards over all episodes)
    cumulative_rewards = np.cumsum(rewards_per_episode)
    
    # Return both rewards per episode and cumulative rewards
    return rewards_per_episode, cumulative_rewards

```
