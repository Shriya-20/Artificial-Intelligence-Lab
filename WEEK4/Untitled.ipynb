```python
import gzip
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance as dist

# Function to parse the TSP dataset and extract city coordinates
def read_tsp_data(file_path):
    with gzip.open(file_path, 'rt') as file:
        lines = file.readlines()
    
    # Extract number of cities
    for line in lines:
        if line.startswith("DIMENSION"):
            num_cities = int(line.split(":")[1])
            break
    
    # Extract coordinates
    coordinates = []
    for line in lines[6:6 + num_cities]:
        coordinates.append([float(x) for x in line.strip().split()[1:]])
    
    return np.array(coordinates)

# Function to compute distance matrix between cities
def compute_distance_matrix(coordinates):
    return dist.squareform(dist.pdist(coordinates))

# Function to solve TSP using different Hill Climbing approaches
def solve_tsp(distance_matrix, method='simple', iterations=1000):
    np.random.seed(69)  # Fix seed for reproducibility
    num_cities = len(distance_matrix)
    best_solution = np.random.permutation(num_cities)  # Initial random solution
    best_distance = sum(distance_matrix[best_solution[i], best_solution[(i+1) % num_cities]] for i in range(num_cities))
    
    for _ in range(iterations):
        i, j = np.random.choice(num_cities, size=2, replace=False)  # Pick two cities to swap
        new_solution = best_solution.copy()
        new_solution[i], new_solution[j] = new_solution[j], new_solution[i]  # Swap two cities
        
        new_distance = sum(distance_matrix[new_solution[k], new_solution[(k+1) % num_cities]] for k in range(num_cities))
        
        if new_distance < best_distance:  # Accept better solution
            best_solution, best_distance = new_solution, new_distance
            
        if method == 'stochastic' and new_distance >= best_distance:  # In stochastic mode, stop if no improvement
            break
        
        if method == 'steepest':  # Steepest ascent: Check all possible swaps
            for x in range(num_cities):
                for y in range(x + 1, num_cities):
                    temp_solution = best_solution.copy()
                    temp_solution[x], temp_solution[y] = temp_solution[y], temp_solution[x]
                    temp_distance = sum(distance_matrix[temp_solution[k], temp_solution[(k+1) % num_cities]] for k in range(num_cities))
                    if temp_distance < best_distance:
                        best_solution, best_distance = temp_solution, temp_distance
    
    return best_solution, best_distance

# Function to visualize the TSP solution
def plot_solution(coordinates, order, title):
    ordered_coordinates = coordinates[order]
    plt.plot(ordered_coordinates[:, 0], ordered_coordinates[:, 1], 'o-', mfc='r')
    plt.title(title)
    plt.show()

# Main execution loop
def run_tsp(file_path):
    coordinates = read_tsp_data(file_path)  # Load data
    distance_matrix = compute_distance_matrix(coordinates)  # Compute distance matrix
    
    # Solve TSP using different methods
    methods = ['simple', 'stochastic', 'steepest']
    for method in methods:
        solution, distance = solve_tsp(distance_matrix, method=method)
        print(f"{method.capitalize()} Hill Climbing: Distance = {distance:.2f}")
        print(f"{solution}")
        plot_solution(coordinates, solution, f'{method.capitalize()} Hill Climbing')

# Example usage:
# run_tsp('path/to/tspfile.tsp.gz')
```

```python
run_tsp('./TSPLIB95/rd100.tsp.gz')
```
